```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
df = pd.read_csv('weather_BE.csv')

# Display first 5 rows
df.head()
```

```python
# Check shape, column names, data types
print(df.shape)
print(df.columns)
df.info()
df.describe()
print("Missing Values:\n", df.isnull().sum())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values (example: fill with mean or drop)
df.fillna(df.mean(numeric_only=True), inplace=True)
# OR
# df.dropna(inplace=True)

# Remove duplicates if any
df.drop_duplicates(inplace=True)

# Optional: Convert 'date' column to datetime if exists
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
```

```python
# Get overall statistical summary
df.describe()
```

```python
# Average temperature by month
if 'date' in df.columns and 'temperature' in df.columns:
    df['month'] = df['date'].dt.month
    print(df.groupby('month')['temperature'].mean())

# Total rainfall per city (if city column exists)
if 'city' in df.columns and 'rainfall' in df.columns:
    print(df.groupby('city')['rainfall'].sum())
```

```python
# Temperature distribution
if 'temperature' in df.columns:
    sns.histplot(df['temperature'], kde=True)
    plt.title('Temperature Distribution')
    plt.xlabel('Temperature')
    plt.ylabel('Frequency')
    plt.show()

# Line plot of temperature over time
if 'date' in df.columns and 'temperature' in df.columns:
    plt.figure(figsize=(12,6))
    plt.plot(df['date'], df['temperature'])
    plt.title('Temperature Over Time')
    plt.xlabel('Date')
    plt.ylabel('Temperature')
    plt.grid(True)
    plt.show()

# Boxplot of humidity by city (if applicable)
if 'city' in df.columns and 'humidity' in df.columns:
    plt.figure(figsize=(10,5))
    sns.boxplot(x='city', y='humidity', data=df)
    plt.title('Humidity by City')
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Between Weather Parameters')
plt.show()
```
