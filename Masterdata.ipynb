```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Predicting Temperature using Humidity and Wind Speed
features = master_df[['Humidity', 'Wind_Speed']]  # or other relevant features
target = master_df['Temperature']

# Check if features and target have enough data points for splitting
if len(features) < 2 or len(target) < 2:
    print("Error: Not enough data points to split into training and testing sets.")
else:
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)  # Added random_state for reproducibility
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predictions
    pred = model.predict(X_test)

    # Plotting Actual vs Predicted
    plt.figure(figsize=(10, 5))
    plt.scatter(y_test, pred)
    plt.xlabel("Actual Temperature")
    plt.ylabel("Predicted Temperature")
    plt.title("Temperature Prediction")
    plt.grid(True)
    plt.show()

    # Evaluate the model (optional)
    from sklearn.metrics import mean_squared_error, r2_score
    print("Mean Squared Error:", mean_squared_error(y_test, pred))
    print("R-squared:", r2_score(y_test, pred))
```

```python
# Rule-based Risk System
def disaster_risk(row):
    # Check if 'Pressure' and 'Rainfall' columns exist in the row
    if 'Pressure' in row and 'Rainfall' in row:
        if row['Pressure'] < 1000 and row['Rainfall'] > 50:
            return '⚠️ High Risk'
        elif row['Pressure'] < 1010:
            return 'Moderate Risk'
        else:
            return 'Low Risk'
    else:
        return 'Unknown'  # Handle cases where columns are missing

master_df['Disaster_Risk'] = master_df.apply(disaster_risk, axis=1)

# Preview
# Check if columns exist before selecting them
if all(col in master_df.columns for col in ['Pressure', 'Rainfall', 'Disaster_Risk']):
    print(master_df[['Pressure', 'Rainfall', 'Disaster_Risk']].head())
else:
    print("Warning: 'Pressure', 'Rainfall', or 'Disaster_Risk' column not found in the DataFrame.")
```
